```python
# numguess 게임 만들어보기

# random은 하나의 class이며 import하면 class 전체를 가지고 옴.
    #그리고 import한 클래스에서 원하는 함수를 사용하기 위해서 "<import한 클래스명>.<함수명>()"이라고 하는 것.
# random을 전부 가지고 오는 것 보다 더 효율적인 방법은 class에서 필요한 함수 하나만 가져오는 법, "from random import randint."
import random


answer = random.randint(1,100+1)
print(answer)
# 100+1을 하는 이유는 slicing할 때 "100"은 포함하지 않고 99까지만 나오기 때문이다.
username = input("Hi there, what's your name? ")
guess = eval(input("Welcome to this game, %s, guess the number please, an integer and nothing else!!! So? Your guess? " % (username)))    

if answer == guess:
# 항상 기준이 되는 값을 왼쪽에 놓는 것이 좋다... 여기서 기준은 answer변수.
    print("Congratulations, " + username + ", you were right! The answer is " + str(answer) + ". You are a genius!")
else:
    print("Hey, %s, you are an idiot, aren't you?" % (username))
```

```python
# Fizzbuzz는 technical interview에서 많이 나오는 질문...

# 1~100의 숫자를 반복해서 확인하며, 3의 배수이면 "fizz"를 출력하고, 5의 배수이면 "buzz"를 출력하고,
    # 15의 배수이면 "fizzbuzz"를 출력하고, 3개 조건 모두 충족하지 않는다면 그 숫자를 그대로 출력한다.
# 조건을 비교하기보다 먼저 1~100의 숫자를 가져와야한다.

for namer in range(1, 100+1):


# 조건 순서는 동시에 3과 5의 배수인 15가 먼저, 그 다음은 3이 5보다 배수의 수가 많으니까 3이 두번째 조건.

if namer % 15 == 0:
        print("fizzbuzz")
    elif namer % 3 == 0:
        print("fizz")
    elif namer % 5 == 0:
        print("buzz")
    else:
        print(str(namer))
```
